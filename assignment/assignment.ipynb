```python
# Q1: Register a student
def register_student(name, marks):
    return {"name": name, "marks": marks}

s1 = register_student("Shahbaz", [85, 92, 78])
print("Registered:", s1)

# Q2: Add multiple students using *args
def add_students(*students):
    return list(students)

s2 = register_student("Ali", [80, 70])
s3 = register_student("Sara", [90, 88])
all_students = add_students(s1, s2, s3)
print("All Students:", all_students)

# Q3: Calculate average marks
def average_marks(student):
    return sum(student["marks"]) / len(student["marks"])

print("Average of Sara:", average_marks(s3))

# Q4: Lambda for grading
grade = lambda avg: "A" if avg >= 90 else "B" if avg >= 75 else "C"
print("Grade for 85:", grade(85))

# Q5: Use map to compute average of all students
def average(student):
    return {
        "name": student["name"],
        "average": sum(student["marks"]) / len(student["marks"])
    }

averages = list(map(average, all_students))
print("Averages:", averages)

# Q6: Use filter to get students with avg >= 85
qualified = list(filter(lambda s: sum(s["marks"]) / len(s["marks"]) >= 85, all_students))
print("Qualified Students:", qualified)

# Q7: Decorator to log function calls
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} completed\n")
        return result
    return wrapper

@logger
def welcome(name):
    print(f"Welcome {name}!")

welcome("Shahbaz")

# Q8: Use recursion to count passed students (avg >= 50)
def count_pass(students):
    if not students:
        return 0
    student = students[0]
    avg = sum(student["marks"]) / len(student["marks"])
    return (1 if avg >= 50 else 0) + count_pass(students[1:])

print("Passed students:", count_pass(all_students))

# Q9: Higher-order function to apply bonus to all marks
def apply_bonus(student, bonus_func):
    student["marks"] = [bonus_func(m) for m in student["marks"]]
    return student

bonus = lambda m: m + 5
updated_student = apply_bonus(register_student("Ahmed", [60, 65]), bonus)
print("Ahmed after bonus:", updated_student)

# Q10: Combine everything in generate_report
def generate_report(student):
    avg = average_marks(student)
    g = grade(avg)
    return {
        "name": student["name"],
        "average": avg,
        "grade": g
    }

report = generate_report(s1)
print("Report for Shahbaz:", report)
```
